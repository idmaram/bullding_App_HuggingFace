```python
# Import required libraries
import torch
from transformers import pipeline

# Check and use GPU if available
device = 0 if torch.cuda.is_available() else -1
print(f"Using device: {'GPU' if torch.cuda.is_available() else 'CPU'}")
```

```python
# Sentiment Analysis

sentiment_pipeline = pipeline("sentiment-analysis", device=device)

sentiment_result = sentiment_pipeline("The weather today is absolutely wonderful, I feel so happy!")
print("Sentiment Analysis:", sentiment_result)
```

```python
# Text Summarization
summarization_pipeline = pipeline("summarization", device=device)

text = """The Amazon rainforest, also known as the lungs of the Earth, plays a crucial role in regulating
          the planet's oxygen and carbon dioxide levels. It covers a vast area of South America and is home
          to a diverse range of wildlife and plant species. However, deforestation and climate change pose
          significant threats to this unique ecosystem."""

summary_result = summarization_pipeline(text, max_length=50, min_length=20, do_sample=False)

print("Summarization:", summary_result)
```

```python
# Named Entity Recognition (NER)
ner_pipeline = pipeline("ner", device=device)
ner_result = ner_pipeline("Albert Einstein was a theoretical physicist who developed the theory of relativity. He lived in USA")
print("Named Entity Recognition:", ner_result)
```

```python
# Question Answering
qa_pipeline = pipeline("question-answering", device=device)
context = "Mount Everest, the highest mountain in the world, is located in the Himalayas on the border between Nepal and Tibet."
question = "Where is Mount Everest located?"
qa_result = qa_pipeline(question=question, context=context)
print("Question Answering:", qa_result)
```

```python
# Text Generation
text_generation_pipeline = pipeline("text-generation", model="gpt2", device=device)
generated_text = text_generation_pipeline("In a futuristic world, robots and humans coexist", max_length=50)
print("Text Generation:", generated_text)
```

```python
# Translation (English to French)
translation_pipeline = pipeline("translation_en_to_fr", device=device)
translation_result = translation_pipeline("The sun is shining brightly, and the birds are singing.")
print("Translation:", translation_result)
```
